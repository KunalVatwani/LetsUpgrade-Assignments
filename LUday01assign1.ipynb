```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
#print(mpl.__version__)
x=np.arange(0,10)
y=x*x
plt.plot(x,y,'ro--')
plt.xlabel('squares of numbers')
plt.ylabel('numbers')
plt.title('square curve')
plt.show()

```
